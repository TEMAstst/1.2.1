Для введенных с клавиатуры положительных целых чисел  и  () определите:

сумму всех целых чисел от a до b;

произведение всех целых чисел от a до b;

среднее арифметическое всех целых чисел от a до b;

среднее геометрическое нечетных чисел от a до b.

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_04_02_13.
#
# Выполнил: Фамилия И.О.
# Группа: !!!
# E-mail: !!!

import math
a = int(input("a ="))
b = int(input("b ="))
n_sum = 0
n_mult = 1
n_count = 0
n_mult_g = 1
n_count_g = 0

for i in range(a, b+1):
    n_sum += i
    n_mult *= i
    n_count += 1
    if i%2 == 1:
        n_mult_g *= i
        n_count_g += 1

n_avg = n_sum / n_count
n_avg_geom = math.pow(n_mult_g, 1/n_count_g)

print("Сумма =", n_sum)
print("Произведение =", n_mult)
print("Среднее арифметическое = {0:.2f}".format(n_avg))
print("Среднее геометрическое нечетных чисел = {0:.2f}".format(n_avg_geom))
# --------------
# Пример вывода:
#
# a = 1
# b = 5
# Сумма = 15
# Произведение = 120
# Среднее арифметическое = 3.00
# Среднее геометрическое нечетных чисел = 2.47
```

пробег лыжника за второй, третий, …, десятый день тренировок; какой суммарный путь он пробежал за первые 10 дней тренировок. При выводе вещественных результатов оставьте один знак после запятой.

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_04_02_14.
#
# Выполнил: Фамилия И.О.
# Группа: !!!
# E-mail: !!!


s = float(input("Пробег за 1-й день (км.) ="))
p = int(input("На сколько увеличивает пробег (%) ="))

total = s
for day in range(2,11):
    s = s + s*p/100
    total += s
    print("Пробег за {}-й день: {:.1f} км.".format(day, s))

print("Суммарный пробег: {:.1f} км.".format(total))

# --------------
# Пример вывода:
#
# Пробег за 1-й день (км.) = 5
# На сколько увеличивает пробег (%) = 10
# Пробег за 2-й день: 5.5 км.
# Пробег за 3-й день: 6.1 км.
# Пробег за 4-й день: 6.7 км.
# Пробег за 5-й день: 7.3 км.
# Пробег за 6-й день: 8.1 км.
# Пробег за 7-й день: 8.9 км.
# Пробег за 8-й день: 9.7 км.
# Пробег за 9-й день: 10.7 км.
# Пробег за 10-й день: 11.8 км.
# Суммарный пробег: 79.7 км.
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_04_02_20.
#
# Выполнил: Цай А.С.
# Группа: АДЭУ-211


n = int(input("n = "))

for i in range(100, 1000): # отобрал все трехзначные числа
  c = i % 10
  b = i // 10 % 10
  a = i // 100 % 10 
  if a + b + c == n:
    print(i, end=' ')


# --------------
# Пример вывода:
#
# n = 3
# 102 111 120 201 210 300
```
